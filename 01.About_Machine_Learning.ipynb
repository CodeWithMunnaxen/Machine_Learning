# Supervised Learning
- Supervised Learning is a type of Machine Learning where the model learns from labeled data. This means the dataset contains both:

```python
# - Input Features (X) → Data given to the model
# - Output Label (Y) → Correct answer
```

# 

# 

# 

# Advantages
- Easy to understand

- High accuracy with good labeled data

- Widely used in real-world applications

- Easy to evaluate performance

# Disadvantages

- Requires large labeled datasets

- Labeling data can be expensive

- Can overfit if data is small

- Performance depends on data quality

# Supervised Learning is a machine learning technique in which a model is trained using labeled data (input and correct output). The model learns the relationship between inputs and outputs and uses that knowledge to make predictions on new data.

# Unsupervised Learning
- Unsupervised Learning is a type of Machine Learning where the model learns from unlabeled data.

```python
# The dataset contains only input features (X) and no target/output label (Y).

# The model tries to find hidden patterns, groups, relationships, or structures in the data by itself
```

| Feature                 | Supervised Learning    | Unsupervised Learning |
| ----------------------- | ---------------------- | --------------------- |
| Data                    | Labeled Data           | Unlabeled Data        |
| Output Label Available? | ✅ Yes                  | ❌ No                  |
| Goal                    | Predict output         | Discover patterns     |
| Example                 | House Price Prediction | Customer Segmentation |


# 

# 

# 

# 

# 

# 

# Advantages

- No labeled data required

- Can discover hidden patterns

- Useful for large datasets

- Helps in data exploration

# Disadvantages

- Results can be difficult to interpret

- No "correct answer" to compare against

- Accuracy is harder to measure

- Sensitive to noise and outliers#

# Scikit-Learn (sklearn) is an open-source Python Machine Learning library that provides tools for data preprocessing, classification, regression, clustering, model selection, and evaluation. It is widely used for building and testing machine learning models quickly and efficiently

```python
from sklearn.linear_model import LinearRegression

# Data
X = [[1000], [1500], [2000], [2500]]
y = [20, 30, 40, 50]

# Create model
model = LinearRegression()

# Train
model.fit(X, y)

# Predict
prediction = model.predict([[1800]])

print(prediction)
```
